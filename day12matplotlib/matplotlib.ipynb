Introduction Matplotlib is a Python plotting library for creating static, interactive, and animated visualizations.

Module: matplotlib.pyplot (commonly imported as plt)

Works well with NumPy, Pandas, and other scientific libraries.


Installation pip install matplotlib


basic usage

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,4,6,8,10,12]
plt.plot(x,y)#basic line chart
plt.show()
```

Matplotlib’s "Object-Oriented" Structure

Figure = the whole sheet of paper where you draw.

Axes = the actual plot area where data is drawn (could be multiple plots in one figure).
Axis = the X-axis or Y-axis within an Axes object (these handle ticks, labels, etc.).



Figure
Created using plt.figure() or via plt.subplots().

Can contain one or more Axes.

Controls overall size, background, layout.

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(6,4),facecolor="lightgray")
#step2 add an axes to the figure
ax=fig.add_subplot(1,1,1)
#step3 plot data inside axes
x=[1,2,3,4]
y=[2,4,6,8]
ax.plot(x,y,marker="o",color="red")
print(type(fig))
#step4 add labels and title
ax.set_title("Basic Line Plot")
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(6,4),facecolor="lightgray")
ax1.plot([1,2,3],[1,4,9],marker="o",color="red")
ax2.plot([1,2,3],[1,2,3],marker="o")
```

plot types

```python
print(x,y)
plt.plot(x,y,color="green",linestyle="--",marker="o")
plt.title("Line Plot")
plt.xlabel("Xasis")
plt.ylabel("laxis")
plt.show()
```

bar chart

```python
catog=["a","b","c"]
val=[10,15,7]
plt.bar(catog,val,color="orange")
plt.title("Bar Chart")
plt.show()
```

```python
plt.barh(catog,val,color="purple")
plt.show()
```

scatter plot

```python
plt.scatter([5,7,8,7],[7,8,5,6],color="red")
plt.title("scattter plot")
plt.show()
```

histogram
graphical representation of frequency distribution of a data set


```python
import numpy as np
data=np.random.randn(1000)
plt.hist(data,bins=30,color="skyblue",edgecolor="black")
plt.title("Histogram")
plt.show()
```

pie chart
shows propotions of a whole as sectors of a circle
angle of each slice is propotiomal to its share of the total

```python
sizes=[40,35,25]
labels=["apples","bananas","cherries"]
colours=["red","yellow","pink"]
plt.pie(sizes,labels=labels,colors=colours,autopct="%1.1f%%")
plt.title("fruit pie chart")
plt.show()
```

```python
import pandas as pd
#sample data frame
data={"month":["jan","feb","mar","apr","may","jun"],
      "sales":[200,250,300,280,350,400]}
df=pd.DataFrame(data)
df.set_index("month",inplace=True)
df["sales"].plot(kind="bar",color="pink",title="Monthly Sales Data")
plt.xlabel("month")
plt.ylabel("sales")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.show()
```
